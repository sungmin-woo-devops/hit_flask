```python
import requests, json, folium

KEY = "D551DDFA-0771-347B-BB9A-CCD8AF3EE46F"
DOMAIN = "localhost"
BASE_URL = "https://api.vworld.kr/req/data"

# 1️⃣ BBOX 생성 (수원시 범위)
center_lng, center_lat = 127.017, 37.267
delta = 0.045
geomFilter = f"BOX({center_lng-delta},{center_lat-delta},{center_lng+delta},{center_lat+delta})"

# 2️⃣ API 호출
params = {
    "service": "data",
    "request": "GetFeature",
    "version": "2.0",
    "data": "LT_L_MOCTLINK",
    "format": "json",
    "geometry": "true",
    "attribute": "true",
    "crs": "EPSG:4326",
    "key": KEY,
    "domain": DOMAIN,
    "geomFilter": geomFilter,
}
res = requests.get(BASE_URL, params=params)
j = res.json()
features = (j.get("response", {})
              .get("result", {})
              .get("featureCollection", {})
              .get("features", []))
print("수집된 링크 수:", len(features))

# 3️⃣ GeoJSON으로 저장
geojson = {
    "type": "FeatureCollection",
    "features": features
}
with open('suwon_links_all.geojson', 'w', encoding='utf-8') as f:
    json.dump(geojson, f, ensure_ascii=False, indent=2)  # 형식 맞춰 저장 :contentReference[oaicite:1]{index=1}

# 4️⃣ Folium으로 시각화
m = folium.Map(location=[center_lat, center_lng], zoom_start=12, tiles='cartodbpositron')
folium.GeoJson(
    geojson,
    name='교통링크',
    style_function=lambda feat: {'color': 'blue','weight': 2,'opacity': 0.7},
    tooltip=folium.GeoJsonTooltip(
        fields=['link_id','road_name','max_spd'],
        aliases=['Link ID','도로명','최고속도'],
        localize=True, labels=True
    )
).add_to(m)
folium.LayerControl().add_to(m)
m.save('suwon_traffic_links_map.html')

with open('suwon_links_all.geojson', 'r', encoding='utf-8') as f:
    geojson_data = json.load(f)

m = folium.Map(location=[37.267, 127.017], zoom_start=12, tiles='cartodbpositron')

folium.GeoJson(
    geojson_data,
    name='교통링크',
    style_function=lambda feature: {
        'color': 'blue',
        'weight': 2,
        'opacity': 0.7
    },
    tooltip=folium.GeoJsonTooltip(fields=['link_id', 'road_name', 'max_spd'],
                                  aliases=['Link ID','도로명','최고속도'])
).add_to(m)

folium.LayerControl().add_to(m)

m.save('suwon_traffic_links.html')
```
